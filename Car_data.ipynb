```python
# Libraries:
import matplotlib.pyplot as plt

import matplotlib
from matplotlib import pyplot as plt

import pandas as pd

import seaborn as sns

import numpy as np
```

# Project
# 1.Load a car CSV Data Set:

```python
df_car = pd.read_csv('Car_data.csv')
df_car
```

# 2.Create new columns for Segment - this is prep for creating a sales trendline by Segment.

```python
# Giving Conditions
conditions = [(df_car['Avg Price'] <= 30000),
    (df_car['Avg Price'] > 30000) & (df_car['Avg Price'] <= 45000),
    (df_car['Avg Price'] > 45000) & (df_car['Avg Price'] <= 75000),
    (df_car['Avg Price'] > 75000)
]

# Defining values for each condition
categories = ['Economy', 'Mid-Range', 'Luxury', 'Ultra-Luxury']

# column based on condition
df_car['Segment'] = pd.Series(np.select(conditions, categories))

df_car
```

```python
# Economy car segment
economy_df = df_car[df_car['Segment'] == 'Economy']
economy_df.drop(columns=['Avg Price'], inplace=True) 
economy_df = economy_df.T
print("Economy Car Segment:") 
print(economy_df)
```

```python
# Mid-Range car segment
mid_range_df = df_car[df_car['Segment'] == 'Mid-Range']
mid_range_df = mid_range_df.T 
print("\nMid-Range Car Segment:")
print(mid_range_df)
```

# 3.Find the percentage change of sales between 2019 & 2020 sales

```python
df_car['Total Sales 2019']=df_car[['Q1 2019','Q2 2019','Q3 2019','Q4 2019']].sum(axis=1)
df_car['Total Sales 2020']=df_car[['Q1 2020','Q2 2020','Q3 2020','Q4 2020']].sum(axis=1)
df_car['Total Sales 2019']
df_car['Total Sales 2020']
```

```python
print('\n Change in Percent 2019 & 2020')

df_car['Total Sales 2019 & 2020'] = (df_car['Total Sales 2019'] -
                                            df_car['Total Sales 2020']  / df_car['Total Sales 2019'] * 100)
df_car['Total Sales 2019 & 2020']
```
