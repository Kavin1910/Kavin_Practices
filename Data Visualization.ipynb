```python
# To see the Correlation between to numerical varibles:
# Correlation na rendukku etha relatability irukkanu pakkarathu

# To extract the Figure size:
plt.figure(figsize = (10,6)) # Here 6 ,6 Represents X and Y Axis

plt.scatter(data = df_iris, x = 'SepalLengthCm',
           y = 'SepalWidthCm',linewidths=1,edgecolors='black', label = 'SepalLengthCm vs SepalWidthCm')
plt.scatter(data = df_iris, x = 'SepalWidthCm',
           y = 'PetalLengthCm',linewidths=1,edgecolors='black', label = 'SepalWidthCm vs PetalLengthCm')
plt.scatter(data = df_iris, x = 'SepalWidthCm',
           y = 'PetalWidthCm',linewidths=1,edgecolors='black', label = 'SepalWidthCm vs PetalWidthCm')
plt.scatter(data = df_iris, x = 'SepalLengthCm',
           y = 'PetalWidthCm',linewidths=1,edgecolors='black', label = 'SepalLengthCm vs PetalWidthCm ')

plt.grid(color = 'c', linestyle = '-', linewidth = .5)

# Definition for Plots:
plt.title('Correlation - Multiple Scatter Plot',loc = 'center')

plt.legend()

plt.show()
```

- As per Karl Pearson Method
- If SepalLengthCm increases, SepalWidthCm also increases
- Means Both SepalLengthCm as well as SepalWidthCm moves in a same direction
- Henceforth, It was a Positive Co-Relation

# Pie chart

- It is a circular graphs divided into sections,displaying the numeric proportion

- It is used to display the Univariate Data

- Each section of the pie plot represents the single cateogry in the data

```python
countries = ['Denmark','Norway','Germany','France','Italy']
population = [5.2,4.1,5.9,7.0,9.6]


# Exploded Pie Chart:
explode = (0, 0, 0.1, 0, 0)

plt.pie(x = population, labels = countries, autopct = '%1.3f%%', shadow = True, radius= 1.5, 
        
       explode = explode, )
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [7.2,3.4]
plt.rcParams['figure.autolayout'] = True


countries = ['Denmark','Norway','Germany','France','Italy']
population = [5.2,4.1,5.9,7.0,9.6]


# Exploded Pie Chart:
# Explode naa thaniya velila varum pirinchu
explodes = (0, 0, 0, 0, 0.3)

'''autopct = '%1.3f%%' - Ithu vanthu syntax - 1 value irukanum 
dot ku aprom ethana values varanum for example %1.2%%'''

plt.pie(x = population, labels = countries, autopct = '%1.3f%%', shadow = True, radius= 0.6,
        frame=False,rotatelabels=False,explode = explodes,labeldistance=1.2)



# Definition for Plots:
plt.title('Population',loc = 'left')

plt.legend(loc='upper right',bbox_to_anchor = (1,1))

# Tight Layout naa box kullaiye varanum nu artham
plt.tight_layout()

# Axis na portions ah equal ah secreate pannu apdingaruthu artham
plt.axis('equal')

plt.show()
```
