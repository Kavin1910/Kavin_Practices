```python
df = pd.DataFrame({'Courses':["Spark","PySpark","Hadoop","Python","PySpark","Spark","Spark"],
    'Fee' :[20000,25000,26000,22000,25000,20000,35000],
    'Duration':['30day','40days','35days','40days','60days','60days','70days'],
    'Discount':[1000,2300,1200,2500,2000,2000,3000]})
df
```

* Groupby Multiple COlumns 

```python
result = df.groupby(['Courses','Fee']).count()
result
```

* Groupby multiple columns on and aggragate on selected column

```python
result = df.groupby(['Courses','Fee'])['Courses'].count()
result
```

* Group multiple columns and aggregate

```python
result = df.groupby(['Courses','Fee'])['Duration'].count()
result
```

* Groupby Multiple Aggragations

```python
result = df.groupby(['Courses','Fee']).aggregate(['min','max','count'])
result
```

* Group by and multiple aggreations on diiferent 

```python
result = df.groupby('Courses').aggregate({'Duration' : 'count',
                                         'Fee' : ['min', 'max']})
result
```
