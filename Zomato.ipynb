# PART â€“ 2 : (Data Analysis)


# 1.Plot the Analysis â€“ What is the Percentage of Restaurants that Accepts Online Orders by using Appropriate Plotting??

```python

online_order_counts = df_zomato['online_order'].value_counts()


percentage_online_order = (online_order_counts / len(df_zomato)) * 100


plt.figure(figsize=(8, 6))
percentage_online_order.plot(kind='bar', color=['skyblue', 'salmon'], edgecolor='black')
plt.title('Percentage of Restaurants Accepting Online Orders')
plt.xlabel('Online Order Availability')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.show()
```

# 2. What is the most common restaurant type and which type has the highest average rating â€“ State the Output with best Plotting??

```python
df_rest_type_rating = df_zomato[['rest_type', 'rate']]

df_rest_type_rating['rate'] = df_rest_type_rating['rate'].str.extract('(\d+\.\d+)').astype(float)

# Most common restaurant type
most_common_rest_type = df_rest_type_rating['rest_type'].mode()[0]

# Restaurant type with the highest average rating
avg_rating_by_rest_type = df_rest_type_rating.groupby('rest_type')['rate'].mean()
highest_avg_rating_type = avg_rating_by_rest_type.idxmax()


plt.figure(figsize=(12, 6))
sns.countplot(x='rest_type', data=df_rest_type_rating, order=df_rest_type_rating['rest_type'].value_counts().index)
plt.title('Most Common Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()


plt.figure(figsize=(12, 6))
sns.barplot(x=avg_rating_by_rest_type.index, y=avg_rating_by_rest_type.values, order=avg_rating_by_rest_type.sort_values(ascending=False).index)
plt.title('Average Rating by Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()

print(f"\nMost Common Restaurant Type: {most_common_rest_type}")
print(f"Restaurant Type with Highest Average Rating: {highest_avg_rating_type}")
```
